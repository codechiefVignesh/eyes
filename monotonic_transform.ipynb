```python
import numpy as np
import pandas as pd
import cv2 as cv
import copy, math, os
import matplotlib.pyplot as plt
```

```python
img_main = cv.imread('C:/Users/vigne/Documents/image_processing/checkerboard_color.png', 1)
```

```python
print(img_main, img_main.shape)
```

```python
plt.imshow(img_main)
```

```python
#monotonic transformation
img_copy = img_main.copy()

min_pixel = np.min(img_copy)
print(f"The minimum pixel value is: {min_pixel}")

img_copy = img_copy - min_pixel
print(img_copy[0])

max_pixel = np.max(img_copy)
print(f"The maxmimum pixel value after removing minimum is : {max_pixel}")

img_copy = img_copy / max_pixel
print(img_copy[0])

scaling_factor = int(input("Enter the scaling factor for the image range"))

img_copy = img_copy * (scaling_factor - 1)
img_copy = np.round(img_copy).astype(np.uint8)
print(img_copy[0])

print(f"The image after doing monotonic transformation by factor {scaling_factor} is ")

plt.figure(figsize = (20, 10))

plt.subplot(121); plt.imshow(img_main); plt.title("The original image")
plt.subplot(122); plt.imshow(img_copy, vmin=0, vmax=255); plt.title("The transformed image")
```
